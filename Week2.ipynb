## Define a function that computes the hypotenuse of a right-angle triangle given its legs by using the Pythagorean theorem

```python
import math

def hypotenuse(leg_1, leg_2):
    # need to wright the block string
    """ returns the hypotenuse of a right triangle using pythagoras' theorem. 
    Args:
        leg_1(float): first leg
        leg_2(float): second leg
    
    Returns:
        float: hypotenuse
        
    >>> math.isclose(hypotenuse(3,4),5)
    True
    >>> hypotenuse(3,4)>5
    False
    
    """
    return math.sqrt(leg_1 ** 2 + leg_2 ** 2)
```

## Define a function that takes as argument a non-negative integer and returns the sum of the individual digits of its argument.

```python
def sum_of_digits(n):
    """Returns the sum of the digits on n

    Args:
        n (int): non-negative integer

    Returns:
        int
        
    >>> sum_of_digits(12345678)
    36
    >>> sum_of_digits(87654321)
    36
    """
    quotient = n
    sum_digits = 0
    while quotient != 0: # peel of the next units place digit in quotient
        #sum_digits = sum_digits + quotient % 10 
        sum_digits += quotient % 10
        quotient = quotient // 10
        
    return sum_digits
```
